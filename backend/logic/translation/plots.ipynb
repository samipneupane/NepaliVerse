```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
```

```python
batches = [0, 300, 920, 1950, 3283, 5653, 12251, 22295, 45019, 68116, 90497, 113337, 136173, 159012, 190680, 204269, 229596, 247405, 265211]
losses = [8.266, 6.098, 5.778, 5.225, 4.821, 4.449, 3.867, 3.448, 2.988, 2.730, 2.566, 2.440, 2.343, 2.265, 2.177, 2.144, 2.090, 2.057, 1.997]

# Fit a smooth line through the data points for better visualization
x_smooth = np.linspace(min(batches), max(batches), 500)
spline = make_interp_spline(batches, losses, k=3)
y_smooth = spline(x_smooth)
```

```python
plt.figure(figsize=(7, 4))
plt.plot(batches, losses, 'o', label="Data Points", markersize=5, color='blue')
plt.plot(x_smooth, y_smooth, label="Smooth Trend Line", color='red')
plt.title("Training Loss over Batches", fontsize=14)
plt.grid(True, which="both", linestyle="--", linewidth=0.4)
plt.show()
```

```python
validation_loss = 2.139
final_training_loss = 1.997 
```

```python
loss_types = ['Final Training Loss', 'Validation Loss']
loss_values = [final_training_loss, validation_loss]

plt.figure(figsize=(6, 4))
plt.bar(loss_types, loss_values, color=['blue', 'orange'])
plt.ylabel("Loss")
plt.title("Final Training vs Validation Loss")

for i, v in enumerate(loss_values):
     plt.text(i, v+0.01, f"{v:.3f}", ha='center', fontweight='bold')

plt.show()
```
